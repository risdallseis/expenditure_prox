```python
cloudw = []
for row in df['text_preproc2']:
    for review in row:
        for word in review:
            cloudw.append(word)
cloudww = " ".join(word for word in cloudw)
```

```python
from wordcloud import WordCloud
from matplotlib.pyplot import figure

figure(figsize=(15, 15), dpi=160)
wordcloud = WordCloud().generate(cloudww)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
